```python
# importamos las librerías que necesitamos

# Tratamiento de datos
# -----------------------------------------------------------------------
import pandas as pd

# Visualización
# ------------------------------------------------------------------------------
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración
# -----------------------------------------------------------------------
pd.set_option('display.max_columns', None) # para poder visualizar todas las columnas de los DataFrames
```

```python
df = pd.read_csv("world_data_full_apply_nulos.csv", index_col=0)
df.head()
```


¿Cómo se distribuye la tasa de natalidad en el conjunto de datos? ¿Hay algún patrón evidente?


```python
# creamos nuestra primera gráfica, en este caso queremos saber como están distribuidas
# para ello usamos el método 'sns.histplot', en este caso le hemos cambiado el color por defecto usando el parámetro 'color' 
sns.histplot(x = "birth_rate", data = df, color = "violet");

# añadimos el método 'plt.xlabel()' para ponerle nombre al eje x
plt.xlabel("tasa natalidad")

# añadimos el método 'plt.ylabel()' para ponerlo nombre al eje y
plt.ylabel("valores");

# Añadimos titulo
plt.title("distribución tasa natalidad")
```

- El eje x representa la tasa de natalidad, lo que significa el número de nacimientos por cada 1,000 personas en la población.
- El eje y muestra la frecuencia o el número de observaciones en cada rango de tasa de natalidad.
- Cada barra en el histograma representa un rango de valores de tasa de natalidad y la altura de la barra indica cuántas observaciones están en ese rango.
- La distribución está sesgada hacia la derecha, indica que la mayoría de los países tienen tasas de natalidad relativamente bajas, pero algunos tienen tasas muy altas.



¿Existe una relación aparente entre la tasa de mortalidad infantil y la esperanza de vida? Proporciona una breve descripción de los patrones que observes.


```python
# modificamos la gráfica anterior añadiendo el parámetro 's' para cambiar el tamaño de los puntos
# además también vamos a cambiar el marcador haciendo uso del parámetro 'marker' 
plt.scatter(x = "infant_mortality", 
            y = "life_expectancy", 
            data = df, 
            c = "purple", 
            s = 3, 
            marker = "^" )

# ponemos los nombres de los ejes
plt.xlabel("Mortalidad infantil")
plt.ylabel("Esperanza de vida")

# ponemos un título usando el método 'plt.title()'
# en este caso lo que hemos añadido es al parámetro 'fontsize' que nos permite cambiar el tamaño del título
plt.title("Relación aparente entre la tasa de mortalidad infantil y la esperanza de vida", fontsize = 8);
```

```python
# Vamos a hacer algunos pequeños cambios en la gráfica creada previamente, como cambiar el color o el tamaño/grosor de los puntos/línea
# creamos una gráfica usando el 'sns.regplot()' para evaluar la relación que existe entre dos variables numéricas
sns.regplot(x = "infant_mortality", 
            y = "life_expectancy", 
            data = df, 
            marker = "d", 
            line_kws = {"color": "black", "linewidth": 1}, # cambiamos el color y el grosor de la linea de tendencia
            scatter_kws = {"color": "teal", "s": 1} # cambiamos el color y el tamaño de los puntos del scaterplot
            )

# cambiamos los nombres de los ejes como hemos estado haciendo hasta ahora
plt.xlabel("Mortalidad infantil", fontsize = 8)
plt.ylabel("Esperanza vida", fontsize = 8)

# ponemos título a la gráfica
plt.title("Relación aparente entre la tasa de mortalidad infantil y la esperanza de vida", fontsize = 8);
```

- Como vemos en la línea de tendencia existe un patrón en el que se aprecia que cuan mayor esperanza de vida tenga el país, menor es la tasa de mortalidad infantil. 
- El eje x nos indica la tasa de mortalidad infantil 
- El eje y la esperanza de vida. 


¿Cuál es el continente con la mayor cantidad de países en el conjunto de datos? Proporciona el porcentaje de países en ese continente.


```python
continentes = df.groupby("continent")["country"].count().reset_index(name= 'continente-pais')
continentes.head()
```

```python
# lo primero que vamos a hacer es definir una lista para definir los colores que queremos para cada una de las categorías. 
# Esta lista debe tener tantos elementos como categorías tenemos en la columna de estudio (en este caso 7)
colores = ["c", "cadetblue", "mediumseagreen", "seagreen", "teal", "turquoise", "lightseagreen"]


# creamos el gráfico basándonos en el DataFrame que hemos creado en el paso anterior
plt.pie("continente-pais", labels= "continent",
        data = continentes, 
        autopct=  '%1.1f%%', 
        colors = colores, 
        textprops={'fontsize': 8}, 
        startangle=90);
```

- Observamos que Africa tiene mayor porcenta ya que tiene mayor números de países con unporcentaje de 27.3%. 
- Europa en cambio tiene menor porcentaje de 24.2% 
- El de menor porcentaje es Norte América con 1%. Debemos tener en cuenta que América está divida por franjas, si fuera América en su totalidad tenemos un porcentaje total de 18%.


¿Cuál es la mediana de la esperanza de vida en Europa? Proporciona un rango intercuartil aproximado y verifica si existen valores atípicos. s


```python
# # empezamos creando el subplot con el número de gráficas deseadas en su interior
# fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (20, 5))

# añadimos un boxplot creado con Seaborn usando el método 'sns.boxplot()'
sns.boxplot(y = "life_expectancy" ,
            x = "continent",
            data = df[df["continent"] == "Europe" ], 
            width = 0.5, 
            color = "turquoise")

plt.title("Mediana esperanza de vida")

plt.xlabel("continente")
        
plt.ylabel("Esperanza de vida")
```

- No se observan valores atípicos. 
- La barra que delimita ambas franjas corresponde con la mediana. 
- La esperanza de vida mayoritaría se observa cogiendo como referencia el tercer quartil en un 75% en la franja entre 76-81 años y el restante 25% entre 81-82 años.
