```python
# importamos las librerías que necesitamos

# Tratamiento de datos
# -----------------------------------------------------------------------
import pandas as pd

# Visualización
# ------------------------------------------------------------------------------
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración
# -----------------------------------------------------------------------
pd.set_option('display.max_columns', None) # para poder visualizar todas las columnas de los DataFrames
```

```python
df = pd.read_csv("files/world_data_full_apply_nulos.csv", index_col=0)
df.head()
```


¿Cómo se distribuye la tasa de natalidad en el conjunto de datos? ¿Hay algún patrón evidente?


```python
# creamos nuestra primera gráfica, en este caso queremos saber como están distribuidas
# para ello usamos el método 'sns.histplot', en este caso le hemos cambiado el color por defecto usando el parámetro 'color' 
sns.histplot(x = "birth_rate", data = df, color = "violet");

# añadimos el método 'plt.xlabel()' para ponerle nombre al eje x
plt.xlabel("tasa natalidad")

# añadimos el método 'plt.ylabel()' para ponerlo nombre al eje y
plt.ylabel("valores");

# Añadimos titulo
plt.title("distribución tasa natalidad")
```

- El eje x representa la tasa de natalidad, lo que significa el número de nacimientos por cada 1,000 personas en la población.
- El eje y muestra la frecuencia o el número de observaciones en cada rango de tasa de natalidad.
- Cada barra en el histograma representa un rango de valores de tasa de natalidad y la altura de la barra indica cuántas observaciones están en ese rango.
- La distribución está sesgada hacia la derecha, indica que la mayoría de los países tienen tasas de natalidad relativamente bajas, pero algunos tienen tasas muy altas.



¿Existe una relación aparente entre la tasa de mortalidad infantil y la esperanza de vida? Proporciona una breve descripción de los patrones que observes.


```python
# modificamos la gráfica anterior añadiendo el parámetro 's' para cambiar el tamaño de los puntos
# además también vamos a cambiar el marcador haciendo uso del parámetro 'marker' 
plt.scatter(x = "infant_mortality", 
            y = "life_expectancy", 
            data = df, 
            c = "purple", 
            s = 3, 
            marker = "^" )

# ponemos los nombres de los ejes
plt.xlabel("Mortalidad infantil")
plt.ylabel("Esperanza de vida")

# ponemos un título usando el método 'plt.title()'
# en este caso lo que hemos añadido es al parámetro 'fontsize' que nos permite cambiar el tamaño del título
plt.title("Relación aparente entre la tasa de mortalidad infantil y la esperanza de vida", fontsize = 8);
```

```python
# Vamos a hacer algunos pequeños cambios en la gráfica creada previamente, como cambiar el color o el tamaño/grosor de los puntos/línea
# creamos una gráfica usando el 'sns.regplot()' para evaluar la relación que existe entre dos variables numéricas
sns.regplot(x = "infant_mortality", 
            y = "life_expectancy", 
            data = df, 
            marker = "d", 
            line_kws = {"color": "black", "linewidth": 1}, # cambiamos el color y el grosor de la linea de tendencia
            scatter_kws = {"color": "teal", "s": 1} # cambiamos el color y el tamaño de los puntos del scaterplot
            )

# cambiamos los nombres de los ejes como hemos estado haciendo hasta ahora
plt.xlabel("Mortalidad infantil", fontsize = 8)
plt.ylabel("Esperanza vida", fontsize = 8)

# ponemos título a la gráfica
plt.title("Relación aparente entre la tasa de mortalidad infantil y la esperanza de vida", fontsize = 8);
```

- Como vemos en la línea de tendencia existe un patrón en el que se aprecia que cuan mayor esperanza de vida tenga el país, menor es la tasa de mortalidad infantil. 
- El eje x nos indica la tasa de mortalidad infantil 
- El eje y la esperanza de vida. 


¿Cuál es el continente con la mayor cantidad de países en el conjunto de datos? Proporciona el porcentaje de países en ese continente.

¿Cuál es la mediana de la esperanza de vida en Europa? Proporciona un rango intercuartil aproximado y verifica si existen valores atípicos. s


```python
# # empezamos creando el subplot con el número de gráficas deseadas en su interior
# fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (20, 5))

# añadimos un boxplot creado con Seaborn usando el método 'sns.boxplot()'
sns.boxplot(y = "life_expectancy" ,
            x = "continent",
            data = df[df["continent"] == "Europe" ], 
            width = 0.5, 
            color = "turquoise")

plt.title("Mediana esperanza de vida")

plt.xlabel("continente")
        
plt.ylabel("Esperanza de vida")
```

- No se observan valores atípicos. 
- La barra que delimita ambas franjas corresponde con la mediana. 
- La esperanza de vida mayoritaría se observa cogiendo como referencia el tercer quartil en un 75% en la franja entre 76-81 años y el restante 25% entre 81-82 años. 


¿Existen relaciones visibles entre la tasa de natalidad y las siguientes variables: tasa de mortalidad infantil, esperanza de vida, densidad de población y población urbana? Usa un subplot. Describe cualquier patrón que encuentres.


```python
# Establecemos el número de filas, columnas que queremos así como el tamaño que vamos a querer que tenga la figura.
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(20, 10))


# Vamos a rellenar la primera (la de la izquierda)
# Al añadir el parámetro ax usamos para referirnos a la gráfica que querramos rellenar. 

sns.regplot(x = "birth_rate", y = "infant_mortality", data = df, 
            marker = "d", 
            line_kws = {"color": "black", "linewidth": 1}, # cambiamos el color y el grosor de la linea de tendencia
            scatter_kws = {"color": "red", "s": 1}, # cambiamos el color y el tamaño de los puntos del scaterplot
            ax = axes[0, 0] # especificamos el eje en el que queremos poner la gráfica. En este caso en el eje 0. 
            )

# para añadir un título usaremos el parámetro '.set_title()' 
axes[0, 0].set_title("Relación entre tasa de natalidad y tasa de esperanza de vida")

# cambiamos el nombre del eje x e y usando el método '.set_xlabel()' y '.set_ylabel()' respectivamente. 
axes[0, 0].set_xlabel("Tasa de natalidad")

# cambiamos el nombre de los ejes usando los métodos
axes[0, 0].set_ylabel("Tasa de mortalidad infantil")

# quitamos la linea de la derecha de la gráfica
axes[0, 0].spines['right'].set_visible(False)

# quitamos la linea de la derecha de la gráfica
axes[0, 0].spines['top'].set_visible(False);

# Repetimos el proceso con las siguientes relaciones cambiando los ejes 

sns.regplot(x = "birth_rate", y = "life_expectancy", data = df, 
            marker = "d", 
            line_kws = {"color": "black", "linewidth": 1}, # cambiamos el color y el grosor de la linea de tendencia
            scatter_kws = {"color": "red", "s": 1}, # cambiamos el color y el tamaño de los puntos del scaterplot
            ax = axes[0, 1] # especificamos el eje en el que queremos poner la gráfica. En este caso en el eje 0. 
            )

# para añadir un título usaremos el parámetro '.set_title()' 
axes[0, 1].set_title("Relación entre tasa de natalidad y esperanza de vida")

# cambiamos el nombre del eje x e y usando el método '.set_xlabel()' y '.set_ylabel()' respectivamente. 
axes[0, 1].set_xlabel("Tasa de natalidad")

# cambiamos el nombre de los ejes usando los métodos
axes[0, 1].set_ylabel("Esperanza de vida")

# quitamos la linea de la derecha de la gráfica
axes[0, 1].spines['right'].set_visible(False)

# quitamos la linea de la derecha de la gráfica
axes[0, 1].spines['top'].set_visible(False);


sns.regplot(x = "birth_rate", y = "density", data = df, 
            marker = "d", 
            line_kws = {"color": "black", "linewidth": 1}, # cambiamos el color y el grosor de la linea de tendencia
            scatter_kws = {"color": "red", "s": 1}, # cambiamos el color y el tamaño de los puntos del scaterplot
            ax = axes[1, 0] # especificamos el eje en el que queremos poner la gráfica. En este caso en el eje 0. 
            )

# para añadir un título usaremos el parámetro '.set_title()' 
axes[1, 0].set_title("Relación entre tasa de natalidad y densidad de población")

# cambiamos el nombre del eje x e y usando el método '.set_xlabel()' y '.set_ylabel()' respectivamente. 
axes[1, 0].set_xlabel("Tasa de natalidad")

# cambiamos el nombre de los ejes usando los métodos
axes[1, 0].set_ylabel("Densidad de población")

# quitamos la linea de la derecha de la gráfica
axes[1, 0].spines['right'].set_visible(False)

# quitamos la linea de la derecha de la gráfica
axes[1, 0].spines['top'].set_visible(False);

sns.regplot(x = "birth_rate", y = "urban_population", data = df, 
            marker = "d", 
            line_kws = {"color": "black", "linewidth": 1}, # cambiamos el color y el grosor de la linea de tendencia
            scatter_kws = {"color": "red", "s": 1}, # cambiamos el color y el tamaño de los puntos del scaterplot
            ax = axes[1, 1] # especificamos el eje en el que queremos poner la gráfica. En este caso en el eje 0. 
            )

# para añadir un título usaremos el parámetro '.set_title()' 
axes[1, 1].set_title("Relación entre tasa de natalidad y población urbana")

# cambiamos el nombre del eje x e y usando el método '.set_xlabel()' y '.set_ylabel()' respectivamente. 
axes[1, 1].set_xlabel("Tasa de natalidad")

# cambiamos el nombre de los ejes usando los métodos
axes[1, 1].set_ylabel("Población urbana")

# quitamos la linea de la derecha de la gráfica
axes[1, 1].spines['right'].set_visible(False)

# quitamos la linea de la derecha de la gráfica
axes[1, 1].spines['top'].set_visible(False);
```

Como podemos ver en las gráficas anteriores: 
- La relación entre tasa de natalidad y mortalidad infantil es 'ascendente' donde vemos que ante un aumento de la tasa de natalidad, la tasa de mortalidad infantil tiende a disminuir.
- La relación entre tasa de natalidad y la esperanza de vida vemos que cuan mayor es la tasa de natalidad, menor es la esperanza de vida en estos países. 
- La relación entre la tasa de natalidad y la densidad de población vemos que no existe una relación clara entre la tasa de natalidad y la densidad de población. Aunque podemos ver en una tasa de natalidad del 10% se agrupan países con una densidad de población 'mayor'. 
- La relación entre la tasa de natalidad y la población urbana apreciamos que no guardan una relación clara. 


¿Cuáles son los 10 países con la mayor área forestal?


```python
# Agrupamos por países y sacamos los países con mayor area forestal y lo ordenamos de mayor a menor
paises = df.groupby("country")["forested_area"].max().sort_values(ascending=False)
paises = paises.reset_index()

# Sacamos sólo 10 países solicitados
top_10paises = paises.head(10)
top_10paises
```

```python
# Ahora vamos a representar y visualizar los resultados en una gráfica 

plt.figure(figsize=(10, 6))

plt.bar(top_10paises['country'],
        top_10paises['forested_area'],
        color='coral')
plt.xlabel('País')
plt.ylabel('Área forestal')
plt.title('Top 10 países con mayor área forestal')
plt.xticks(rotation=45, ha='right')
```
